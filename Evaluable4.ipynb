```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('covid19_tweets.csv')
df.head(10)
```

```python
df = pd.DataFrame(
    pd.read_csv("covid19_tweets.csv")[["user_followers", "user_friends"]])
df.head()
```

```python
plt.figure(figsize=(5, 7))
sns.boxplot(data=df, orient="vertical", showfliers=False)
plt.title("Diagrama con Cajas y Bigotes")
plt.show()
```

```python
plt.figure(figsize=(7, 3))
plt.subplot(1, 2, 1)
sns.histplot(df["user_followers"], bins=10, kde=True)
plt.title("Histograma de user_followers")

plt.subplot(1, 2, 2)
sns.histplot(df["user_friends"], bins=10, kde=True)
plt.title("Histograma de user_friends")

plt.tight_layout()
plt.show()
```
